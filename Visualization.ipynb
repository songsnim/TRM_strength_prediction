```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
feature_names_label = ['빔 폭',
                '빔 높이',
                '인장철근 깊이',
                '압축철근 깊이',
                '철근콘크리트 길이',
                '빔 순지간',
                '하중재하길이',
                '콘크리트 28일 압축강도',
                '인장철근 단면적',
                '인장철근 항복강도',
                '압축철근 단면적',
                '압축철근 항복강도',
                'Carbon',
                'PBO',
                'Glass',
                'CFRP',
                'Basalt',
                '섬유 인장강도',
                '보강 텍스타일 단면적',
                '보강 텍스타일 레이어 수',
                '종방향 메쉬 크기',
                '횡방향 메쉬 크기',
                '전단보강 여부','항복강도']
```

```python
path = 'data/raw_data.csv'
dataset = pd.read_csv(path)
print('dataset 내 null 갯수 :',dataset.isnull().sum().sum())
dataset.columns = feature_names_label
dataset
```

```python
materials = ['PBO','Carbon','Glass','Basalt','CFRP']
material_lenlist = []
for material in materials:
    material_lenlist.append(len(np.where(dataset[material]==1)[0]))

plt.figure(figsize=(12,5),dpi=100)
plt.title('Material counts',fontsize=14)
plt.bar(materials,material_lenlist)
```

```python
# %matplotlib inline
corr = dataset.corr(method='pearson')
fig, ax = plt.subplots(figsize=(15,15),dpi=100)
im = ax.imshow(corr, cmap='Reds')

# Show all ticks and label them with the respective list entries
ax.set_xticks(np.arange(len(dataset.columns)), labels=dataset.columns)
ax.set_yticks(np.arange(len(dataset.columns)), labels=dataset.columns)

# Rotate the tick labels and set their alignment.
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

# Loop over data dimensions and create text annotations.
for i in range(len(dataset.columns)):
    for j in range(len(dataset.columns)):
        text = ax.text(j, i, round(np.array(corr)[i, j],2),
                       ha="center", va="center")

ax.set_title("Pearson correlation matrix",fontsize=15)
fig.tight_layout()
plt.show()
```

```python
x = np.arange(-4,4,0.01)

def elu(list):
    result = []
    for x in list:
        if x>=0:
            y=x
        elif x<0:
            y=np.exp(x)-1
        result.append(y)
    return result

plt.figure(figsize=(12,8),dpi=100)
plt.title('ELU activation function', fontsize=20)
plt.xlabel('input value',fontsize=20)
plt.ylabel('activation value',fontsize=20)
plt.plot([0]*500,np.arange(-1,4,0.01),c='grey',ls='--')
plt.plot(x,[0]*800, c='grey',ls='--')
plt.plot(x,elu(x), c='black')
```
